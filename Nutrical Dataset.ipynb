```python
#Comparitive Bar Graph for Iron and calcium
iron = df.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = df.groupby('Category')['Calcium (% Daily Value)'].mean()
plt.figure(figsize=(10, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values

Y_axis = np.arange(len(category))

plt.barh(Y_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'blue')
plt.barh(Y_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'green')

plt.yticks(Y_axis, category)
plt.ylabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.legend()
plt.show()
```

```python
#2,000 calories per day

df['Calories%DV']= (df['Calories']*100)/2000 #Calculating intake of %Daily Value of calories 

#Intake of %Daily Value of calories item wise- maximum %DV of calories absorb noted with Chicken McNuggets

df[['Item','Calories','Calories%DV']].groupby('Item').max().sort_values(by='Calories%DV', ascending=False).head(25)
```

```python
#Sugar 50 grams a day
df['Sugar%DV']= (df['Sugars']*100)/50 #Calculating intake of %Daily Value of Sugars 

#5-20 Rule: 5% DV  Lower in saturated fat, sodium, and added sugars
da=df[((df['Total Fat (% Daily Value)']<=5) & (df['Saturated Fat (% Daily Value)']<=5) & (df['Cholesterol (% Daily Value)']<=5) & (df['Sodium (% Daily Value)']<=5) & (df['Sugar%DV']<=5))] 
da[['Category','Item','Total Fat (% Daily Value)','Saturated Fat (% Daily Value)','Cholesterol (% Daily Value)','Sodium (% Daily Value)','Sugar%DV']]
```

```python
#Bar Graph for HEALTHY FOOD
plt.figure (figsize=(10, 4), dpi=100) #input plot size
df['5%DV']=((df['Total Fat (% Daily Value)']<=5) & (df['Saturated Fat (% Daily Value)']<=5) & (df['Cholesterol (% Daily Value)']<=5) & (df['Sodium (% Daily Value)']<=5) & (df['Sugar%DV']<=5)) 
menu_ldv = df.groupby('Category')['5%DV'].sum().sort_values(ascending= False)
menu_ldv.plot.barh(color = ['blue','green','green','green','green','green','green','green','green'])
plt.title("5-20 Rule 5% DV Lower in saturated fat, sodium, and added sugars- HEALTHY FOOD")
plt.ylabel("Categories")
plt.show()
```

```python
#Protein 50 grams a day

df['Protein%DV']= (df['Protein']*100)/50 #Calculating intake of %Daily Value of Protein 

#5-20 Rule: 20% DV Higher in dietary fiber, vitamins , calcium, iron, and potassium.

da= df[(df['Dietary Fiber (% Daily Value)']>=20) | (df['Vitamin A (% Daily Value)']>=20) | (df['Vitamin C (% Daily Value)']>=20) |
       (df['Calcium (% Daily Value)']>=20) | (df['Iron (% Daily Value)']>=20)]
da[['Category','Item','Dietary Fiber (% Daily Value)','Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)','Iron (% Daily Value)']]
```

```python
#Bar Graph for HEALTHY FOOD
plt.figure (figsize=(10, 4), dpi=100) #input plot size
df['20%DV']= ((df['Dietary Fiber (% Daily Value)']>=20) |  (df['Vitamin A (% Daily Value)']>=20) | (df['Vitamin C (% Daily Value)']>=20) |
       (df['Calcium (% Daily Value)']>=20) | (df['Iron (% Daily Value)']>=20))
menu_ldv = df.groupby('Category')['20%DV'].sum().sort_values(ascending= False)
menu_ldv.plot.barh(color = ['blue','green','green','green','green','green','green','green','green'])
plt.title("5-20 Rule 20% DV Higher in dietary fiber, vitamins , calcium, iron, and potassium.- HEALTHY FOOD")
plt.ylabel("Categories")
plt.show()
```
